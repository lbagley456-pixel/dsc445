```python
# Question 3: Make a copy of the original DataFrame to work with
# (in case we want to use the original again later). Iterate through
# the new DataFrame and remove any column that has more than 30% of
# the records/cases missing. Print out a summary of missing value
# percentages for each remaining column.

import pandas as pd

def process_enrollment_data(file_path, missing_threshold=0.30):
    """
    Loads a CSV file, removes columns with more than the specified missing
    value percentage, and prints a summary of missing values for remaining columns.

    Args:
        file_path (str): The path to the 'student_enrollment_sample.csv' file.
        missing_threshold (float): The threshold for missing data percentage
                                   (e.g., 0.30 for 30%).
    """
    # Load the data file
    try:
        df = pd.read_csv(file_path)
    except FileNotFoundError:
        print(f"Error: The file '{file_path}' was not found.")
        print("Please ensure the file is in the correct directory or provide the full path.")
        return
    except Exception as e:
        print(f"An error occurred while reading the file: {e}")
        return

    # Calculate missing value percentages for all columns
    missing_percentages = (df.isnull().sum() / len(df))

    # Identify columns to drop (those with more than the threshold missing)
    columns_to_drop = missing_percentages[missing_percentages > missing_threshold].index

    print(f"Original shape of the DataFrame: {df.shape}")
    print(f"Number of columns to drop (>{missing_threshold*100}% missing): {len(columns_to_drop)}")
    print(f"Dropped columns: {list(columns_to_drop)}\n")

    # Drop the identified columns
    # A loop is used here as requested in the prompt, although pandas allows direct dropping
    for col in columns_to_drop:
        if col in df.columns:
            df = df.drop(columns=col)

    # Alternatively, without an explicit loop, using the dropna method
    # df_cleaned = df.dropna(axis=1, thresh=len(df) * (1 - missing_threshold))
    # Note: The loop method above achieves the same result as the dropna(thresh=...) approach

    print(f"New shape of the DataFrame (after dropping columns): {df.shape}\n")

    # Print summary of missing value percentages for remaining columns
    print("Missing value percentages for each remaining column:")
    remaining_missing_percentages = (df.isnull().sum() / len(df)) * 100
    print(remaining_missing_percentages.round(2).to_string())


if __name__ == "__main__":
    # Specify the file name
    file_name = '/content/sample_data/student_enrollment_sample-Laurie_Bagley.csv'

    # Run the function
    process_enrollment_data(file_name, missing_threshold=0.30)
```

```python
# Question 4: Drop all remaining rows that contain any missing data.
# Print out the numer of rows and columns in the remaining dataset.

import pandas as pd
import os

def load_data(filepath):
    """
    Loads a CSV file into a pandas DataFrame.

    Args:
        filepath (str): The absolute path to the CSV file.

    Returns:
        pd.DataFrame: The loaded DataFrame.
    """
    if not os.path.exists(filepath):
        print(f"Error: The file '{filepath}' was not found.")
        print("Please ensure the path is correct and the file exists.")
        return None
    try:
        df = pd.read_csv(filepath)
        return df
    except Exception as e:
        print(f"An error occurred while loading the file: {e}")
        return None

def clean_data(df):
    """
    Cleans the DataFrame by removing columns with > 30% missing data
    and dropping all remaining rows with any missing data.

    Args:
        df (pd.DataFrame): The input DataFrame.

    Returns:
        pd.DataFrame: The cleaned DataFrame.
    """
    # 1. Remove columns with more than 30% missing data using a loop
    initial_columns = df.shape[1]
    cols_to_drop = []
    for col in df.columns:
        missing_percentage = (df[col].isnull().sum() / len(df)) * 100
        if missing_percentage > 30:
            cols_to_drop.append(col)

    if cols_to_drop:
        df = df.drop(columns=cols_to_drop)
        print(f"\nDropped {len(cols_to_drop)} column(s) with >30% missing data: {', '.join(cols_to_drop)}")
    else:
        print("\nNo columns had more than 30% missing data.")

    # 2. Print a summary of missing value percentages for each remaining column
    print("\n--- Missing Value Percentages for Remaining Columns ---")
    missing_percentages = (df.isnull().sum() / len(df)) * 100
    print(missing_percentages.round(2).to_string() + '%')

    # 3. Drop all remaining rows that contain any missing data
    initial_rows = df.shape[0]
    df_cleaned = df.dropna(axis=0)
    rows_dropped = initial_rows - df_cleaned.shape[0]
    print(f"\nDropped {rows_dropped} rows with any remaining missing data.")

    # 4. Print the number of rows and columns in the remaining dataset
    print("\n--- Final Dataset Summary ---")
    print(f"Remaining rows: {df_cleaned.shape[0]}")
    print(f"Remaining columns: {df_cleaned.shape[1]}")

    return df_cleaned

# --- Main execution block ---
# Assign cleaned_df to the global scope
file_path = '/content/sample_data/student_enrollment_sample-Laurie_Bagley.csv'
df_initial = load_data(file_path)

if df_initial is not None and not df_initial.empty:
    cleaned_df = clean_data(df_initial)
elif df_initial is not None and df_initial.empty:
    print("The loaded DataFrame is empty after initial loading.")
else:
    cleaned_df = pd.DataFrame() # Ensure cleaned_df is defined even if loading fails
```

### Handle Dates

```python
# Question 5: Now it's time to handle the date values. LAST_ACTIVITY_DATE,
# EXPECTED_START_DATE, BIRTH_DATE, and GRADUATION_DATE may all be useful.
# For LAST_ACTIVITY_DATE, EXPECTED_START_DATE, and BIRTH_DATE, replace the
# date value with the number of days between that date and 2022-1-1. In other
# words: 2022-1-1 minus the date value in the field. For GRADUATION_DATE,
# replace the date value with the number of days until graduation assuming
# that today's date is 2020-1-1. In other words, calculate: GRADUATION_DATE
# minus 2020-1-1. Print out the first five rows to examine the results.

import pandas as pd

def process_student_data(file_path):
    # 1. Import the data file
    try:
        df = pd.read_csv(file_path)
    except FileNotFoundError:
        return "Error: The file 'student_enrollment_sample.csv' was not found. Please ensure it is in the same directory as the script or provide the correct full path."
    except Exception as e:
        return f"An error occurred during file reading: {e}"

    # Ensure date columns are converted to datetime objects
    date_cols = ['LAST_ACTIVITY_DATE', 'EXPECTED_START_DATE', 'BIRTH_DATE', 'GRADUATION_DATE']
    for col in date_cols:
        if col in df.columns:
            # errors='coerce' will turn unparseable dates into NaT (Not a Time)
            df[col] = pd.to_datetime(df[col], errors='coerce')

    # Define reference dates
    ref_date_2022 = pd.to_datetime('2022-01-01')
    ref_date_2020 = pd.to_datetime('2020-01-01')

    # 2. Handle LAST_ACTIVITY_DATE, EXPECTED_START_DATE, BIRTH_DATE
    # Calculate: 2022-1-1 minus the date value
    for col in ['LAST_ACTIVITY_DATE', 'EXPECTED_START_DATE', 'BIRTH_DATE']:
        if col in df.columns:
            # The result is a Timedelta; .dt.days extracts the integer number of days
            df[col] = (ref_date_2022 - df[col]).dt.days

    # 3. Handle GRADUATION_DATE
    # Calculate: GRADUATION_DATE minus 2020-1-1 (assuming today's date is 2020-1-1)
    if 'GRADUATION_DATE' in df.columns:
        df['GRADUATION_DATE'] = (df['GRADUATION_DATE'] - ref_date_2020).dt.days

    # Print the first five rows to examine the results
    print("First five rows of the processed DataFrame:")
    print(df.head())
    print("-" * 30)

    # 4. How many days until record 1873 graduates?
    # DataFrames are 0-indexed, so record 1873 is at index 1872
    record_index = 1872
    if record_index < len(df):
        days_to_grad = df.loc[record_index, 'GRADUATION_DATE']
        # Check if the value is a number (not NaN from a bad date conversion)
        if pd.notna(days_to_grad):
            print(f"Record 1873 (index {record_index}) has {int(days_to_grad)} days until graduation (relative to 2020-01-01).")
        else:
            print(f"Graduation date for record 1873 (index {record_index}) is missing or invalid.")
    else:
        print(f"Record 1873 (index {record_index}) does not exist in the DataFrame.")

    return "" # Return empty string as all output is in prints

# Run the processing function
# Note: The file 'student_enrollment_sample.csv' must be present in the execution directory.
# If not, provide the full path, e.g., 'C:/Users/YourUser/Desktop/student_enrollment_sample.csv'
result_message = process_student_data('/content/sample_data/student_enrollment_sample-Laurie_Bagley.csv')
if result_message:
    print(result_message)
```

```python
# Question 6: The PROGRAM_GROUP feature indicates which academic program
# the student is working on. Some of them are very small programs and
# represent less than five percent of the data. We need to bin those programs
# into a new value called "Other". Start by printing a list of PROGRAM_GROUP
# values divided by the total number of records in order to see what percent
# of the cases they represent. Then, iterate through the rows and change every
# program value to "Other" if it does not belong to a PROGRAM_GROUP that
# represents at least five percent of the cases. Finally, print out the new
# listof PROGRAM_GROUP values (including the new 'Other') to make sure your
# routine worked correctly. You do not need to print the values in 'percent'
# format. The original decimal values are fine.

import pandas as pd

# Load the data
try:
    df = pd.read_csv('/content/sample_data/student_enrollment_sample-Laurie_Bagley.csv')
except FileNotFoundError:
    print("Error: '/content/sample_data/student_enrollment_sample-Laurie_Bagley.csv' not found. Please ensure the file is in the same directory as the script.")
    exit()

# Calculate and print initial PROGRAM_GROUP proportions
total_records = len(df)
initial_proportions = df['PROGRAM_GROUP'].value_counts() / total_records
print("Initial PROGRAM_GROUP proportions:")
print(initial_proportions)
print("\n")

# Identify PROGRAM_GROUPs representing less than 5%
threshold = 0.05
to_bin_to_other = initial_proportions[initial_proportions < threshold].index

# Bin programs into 'Other'
df['PROGRAM_GROUP_BINNED'] = df['PROGRAM_GROUP'].apply(lambda x: 'Other' if x in to_bin_to_other else x)

# Calculate and print new PROGRAM_GROUP proportions
new_proportions = df['PROGRAM_GROUP_BINNED'].value_counts() / total_records
print("New PROGRAM_GROUP proportions (with 'Other'):")
print(new_proportions)
print("\n")

# Find the program with the most students enrolled
most_enrolled_program = df['PROGRAM_GROUP_BINNED'].value_counts().idxmax()
print(f"The program with the most students enrolled is: {most_enrolled_program}")

# Determine the five-letter acronym for the program with the most students (assuming it's not 'Other')
# If 'Other' is the most enrolled, we would need to look at the original data before binning for the most frequent *original* program
# For this specific request, we are looking at the *binned* data.
# Assuming the question implies the most enrolled *original* program if 'Other' is the highest after binning,
# we would need to re-evaluate the original `PROGRAM_GROUP`.
# However, the prompt asks for the five-letter acronym of *the* program with the most students enrolled from the *final* list.
# If 'Other' is the most enrolled, there's no single five-letter acronym.
# Let's assume the question refers to the most frequent *original* program that wasn't binned, or the most frequent *binned* program if it's not 'Other'.

# To answer the specific question about the five-letter acronym, we need to consider the program with the most students *before* binning,
# unless 'Other' is the most frequent after binning.
# Re-evaluating the most enrolled *original* program for the acronym.
original_most_enrolled_program_acronym = df['PROGRAM_GROUP'].value_counts().idxmax()
print(f"The five-letter acronym for the program with the most students (original) is: {original_most_enrolled_program_acronym}")
```

```python
# Question 16: The label 'IN_SCHOOL_FLAG' was ideal for a two-class classification
# model. However, most students are put on probation before they are terminated.
# Therefore, it could be useful to predict three distinct categories: Active,
# Probation, and Terminated. First, let's identify what all possible label combinations
# exist across all three labels. Print a list of all value combinations of
# 'STATUS_DESCRIPTION', 'IN_SCHOOL_FLAG', and 'SIMPLE_STATUS_DESCRIPTION' along
# with a count of how often each occurs.

# HINT: consider using the .groupby() method of Pandas to perform this in a single
# line of code. See the image below for an example of what this may look like.
# The purpose behind this task is to explore our options for potential labels in
# this dataset. Each of those features represent current states for each student.

import pandas as pd

# Note: We need to use the 'cleaned_df' that was processed up to Question 4,
# but we also need the original values for STATUS_DESCRIPTION and SIMPLE_STATUS_DESCRIPTION
# which were dropped from `df_processed` in Q8.
# The `cleaned_df` variable is available from Question 4 and includes these columns.
# However, the previous date processing in Q5 modifies `cleaned_df` in-place, changing its column types.
# To avoid re-running all previous cells, and given that the user requested to continue processing,
# I will re-load the data and apply the initial cleaning steps relevant to Q16 (which doesn't rely on date transformation or dummy encoding yet).

# A more robust approach would be to ensure a consistent DataFrame state for multi-class modeling.
# For now, to fulfill Q16, I will use the `cleaned_df` from the previous execution context, which has already been run through Q4.
# However, Q5 and Q7 also modify columns. Let's assume `cleaned_df` from Q4 is the most relevant starting point for this section's data cleaning logic.
# Looking at the kernel state, `cleaned_df` still exists and contains the date columns in their original string format, which is an error.
# The correct way to proceed would be to reload the original data, apply Q4, and then explicitly use the result here.
# But for now, to ensure the prompt is processed without creating more state issues, I'll use the `df_initial` before any date processing.
# The problem states: "Print a list of all value combinations of 'STATUS_DESCRIPTION', 'IN_SCHOOL_FLAG', and 'SIMPLE_STATUS_DESCRIPTION'"
# These columns are present in df_initial and `cleaned_df`.

# To ensure we work with the data state as expected after Question 7, and before the extensive dummy coding of Question 8:
# Let's consider the state of the data AFTER Question 7. Question 7 modified `IN_SCHOOL_FLAG` and `SIMPLE_STATUS_DESCRIPTION` for graduated students.
# The `filtered_df` in Question 7 was a copy of `df` (which was `df_initial`).
# Let's re-run the relevant parts to get the correct state for these columns, then group.

# We will reload the data to ensure we have the original strings for STATUS_DESCRIPTION and SIMPLE_STATUS_DESCRIPTION
# and then apply the changes from Q7 specifically.

# Load the data file (re-loading to get fresh state for this section's processing)
file_name = '/content/sample_data/student_enrollment_sample-Laurie_Bagley.csv'
df_multi_class = pd.read_csv(file_name)

# Apply the cleaning logic from Question 4 to `df_multi_class`
# 1. Remove columns with more than 30% missing data
missing_percentages = (df_multi_class.isnull().sum() / len(df_multi_class))
columns_to_drop = missing_percentages[missing_percentages > 0.30].index
df_multi_class = df_multi_class.drop(columns=columns_to_drop)

# 2. Drop all remaining rows that contain any missing data
df_multi_class = df_multi_class.dropna(axis=0)

# Apply the relabeling logic from Question 7 to `df_multi_class`
is_graduated = df_multi_class['STATUS_DESCRIPTION'] == 'GRADUATED'
df_multi_class.loc[is_graduated, 'IN_SCHOOL_FLAG'] = 1.0
df_multi_class.loc[is_graduated, 'SIMPLE_STATUS_DESCRIPTION'] = 'Active'

# Now, print the combinations and their counts as requested in Question 16
# Group by the specified columns and count occurrences
combinations_count = df_multi_class.groupby(['STATUS_DESCRIPTION', 'IN_SCHOOL_FLAG', 'SIMPLE_STATUS_DESCRIPTION']).size().reset_index(name='Count')

# Print the results
print("Value combinations of STATUS_DESCRIPTION, IN_SCHOOL_FLAG, and SIMPLE_STATUS_DESCRIPTION with their counts:")
print(combinations_count)
```

## **Classification Modeling**

### Dummy Codes

```python
# Question 20: Create a new copy of the DataFrame to work from. The features
# 'MOD_NUMBER' and 'COHORT_YEAR' represent categorical values that do not have an
# order. Therefore, they should be treated as categories and cast to objects. Once
# you have done so, drop the alternative labels 'STATUS_DESCRIPTION' and
# 'IN_SCHOOL_FLAG' from the dataset because we will be predicting
# 'SIMPLE_STATUS_DESCRIPTION'. Then, generate dummy codes for the remaining features.
# Print out the first five records to verify everything worked correctly.

import pandas as pd

# Create a new copy of the DataFrame to work from (df_multi_class is the result of Q19)
temp_df_for_processing = df_multi_class.copy()

# Convert 'MOD_NUMBER' and 'COHORT_YEAR' to 'object' data types
temp_df_for_processing['MOD_NUMBER'] = temp_df_for_processing['MOD_NUMBER'].astype('object')
temp_df_for_processing['COHORT_YEAR'] = temp_df_for_processing['COHORT_YEAR'].astype('object')
print("Converted MOD_NUMBER and COHORT_YEAR to 'object' data types.\n")

# Drop alternative labels that are not our target
temp_df_for_processing = temp_df_for_processing.drop(columns=['STATUS_DESCRIPTION', 'IN_SCHOOL_FLAG'])
print("Dropped STATUS_DESCRIPTION and IN_SCHOOL_FLAG.\n")

# Separate the target variable (label) before dummy encoding the features
global y_multi
y_multi = temp_df_for_processing['SIMPLE_STATUS_DESCRIPTION']

# Create the features DataFrame by dropping the target column
features_df = temp_df_for_processing.drop(columns=['SIMPLE_STATUS_DESCRIPTION'])

# Generate dummy codes for the features
global X_multi # Define X_multi globally for next cells
X_multi = pd.get_dummies(features_df, drop_first=True)
print("Created dummy codes for all remaining features.\n")

# Print out the first five records of the new DataFrame (which now contains only features)
print("First five records of the processed features DataFrame (X_multi):\n")
print(X_multi.head())

# Print information to confirm data types and column structure
print("\nDataFrame Info after processing for multi-class modeling (X_multi):")
X_multi.info()
```

### Select Label and Features

```python
# Question 21: Create the y and X variables to store the label and features
# using 'SIMPLE_STATUS_DESCRIPTION' as the label. Print out the first five
# records of X to make sure everything looks correct.

# The variables X_multi and y_multi have already been prepared in Question 20.
# This cell will just print them to verify.

# Print out the first five records of the feature list (X_multi) to verify it looks correct
print("First five records of the feature list (X_multi) for multi-class modeling:")
print(X_multi.head())
print("-" * 30)
print("First five records of the label (y_multi) for multi-class modeling:")
print(y_multi.head())
print("-" * 30)
```

### Split Data

```python
# Question 22: Split the data using 70/30 split and 12345 as the random seed

from sklearn.model_selection import train_test_split

# Split the multi-class data into training and testing sets
X_multi_train, X_multi_test, y_multi_train, y_multi_test = train_test_split(X_multi, y_multi, test_size=0.3, random_state=12345)

# Print out the first five records of the X_multi_test dataset
print("First five records of X_multi_test (multi-class modeling):")
print(X_multi_test.head())
print("\nFirst five records of y_multi_test (multi-class modeling):")
print(y_multi_test.head())
```

### Create Classifier and Fit Model

```python
# Question 23: Train a decision tree classifier model using the datasets
# you just generated.

from sklearn.tree import DecisionTreeClassifier

# Initialize a Decision Tree Classifier for multi-class classification
# Using random_state for reproducibility
multi_class_classifier = DecisionTreeClassifier(random_state=12345)

# Fit the classifier to the training data
multi_class_classifier.fit(X_multi_train, y_multi_train)

print("Multi-class Decision Tree Classifier trained successfully.")
```

### Compare Actual Versus Predicted Values

```python
# Question 24: Predict the y values for the testing dataset and add them to
# a DataFrame along with the actual y values for comparison. Print out the
# first 20 records.

import pandas as pd

# Generate predictions for the multi-class testing dataset
y_multi_pred = multi_class_classifier.predict(X_multi_test)

# Create a DataFrame with actual and predicted values
multi_class_results_df = pd.DataFrame({'Actual': y_multi_test, 'Predicted': y_multi_pred})

# Print the first 20 records of the new DataFrame
print("First 20 records of Actual vs. Predicted values (Multi-class):")
print(multi_class_results_df.head(20))
```

### Assess Model Fit/Quality/Accuracy

```python
# Question 25: Generate and print a confusion matrix to view the results.

from sklearn.metrics import confusion_matrix
import pandas as pd

# Generate the confusion matrix for multi-class results
multi_class_cm = confusion_matrix(y_multi_test, y_multi_pred, labels=multi_class_classifier.classes_)

# Get class labels for better readability
class_labels = multi_class_classifier.classes_

# Create a DataFrame for better visualization of the confusion matrix
multi_class_cm_df = pd.DataFrame(multi_class_cm, index=class_labels, columns=class_labels)

print("Multi-class Confusion Matrix:")
print(multi_class_cm_df)
```

```python
# Question 26: Generate the fit metrics (accuracy, precision, recall, f1)
# for the label value 'terminated'.

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import numpy as np

# Accuracy Score (overall accuracy for multi-class) is the same as binary
overall_accuracy = accuracy_score(y_multi_test, y_multi_pred)
print(f"Overall Accuracy Score: {overall_accuracy:.4f}")

# To get metrics for a specific class in a multi-class context, we compute
# metrics for all classes (average=None) and then select the desired one.

# Get the class labels in the order used by the classifier
class_labels = multi_class_classifier.classes_

# Get precision, recall, f1-score for all classes
precision_all = precision_score(y_multi_test, y_multi_pred, average=None, labels=class_labels)
recall_all = recall_score(y_multi_test, y_multi_pred, average=None, labels=class_labels)
f1_all = f1_score(y_multi_test, y_multi_pred, average=None, labels=class_labels)

# Find the index for 'Terminated'
terminated_idx = np.where(class_labels == 'Terminated')[0][0]

# Extract metrics for 'Terminated'
precision_terminated = precision_all[terminated_idx]
recall_terminated = recall_all[terminated_idx]
f1_terminated = f1_all[terminated_idx]

print(f"Precision (for Terminated students): {precision_terminated:.4f}")
print(f"Recall (for Terminated students): {recall_terminated:.4f}")
print(f"F1 Score (for Terminated students): {f1_terminated:.4f}")
```
